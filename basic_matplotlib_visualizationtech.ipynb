```python
import pandas as pd
from google.colab import files
uploaded = files.upload()
```

```python
import io
df = pd.read_csv(io.BytesIO(uploaded['salaries_by_college_major (1).csv']))
import numpy as np
```

```python
df = df.dropna()
df.head()
```

```python
df = df.drop(columns=['Mid-Career 10th Percentile Salary','Mid-Career 90th Percentile Salary'])
df.head()
```

```python
df = df.drop(columns=['Group'])
df.head()
```

```python
#changing the values of a column to the index (row values)
ddf = df.set_index('Undergraduate Major')
ddf
```

```python
import matplotlib.pyplot as plt
```

```python
x_value = df['Undergraduate Major']
```

```python
y_value = df['Starting Median Salary']
```

```python
plt.plot(x_value, y_value)
```

```python
x_axis = ddf.index
y_axis = ddf.columns
plt.plot(x_value, y_value)
```

```python
x = np.arange(len(ddf.index))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0
x
```

```python
fig, ax = plt.subplots(figsize=(100,5))


for attribute, measurement in ddf.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Dollars')
ax.set_title('Start vs. Mid-Career Salaries')
ax.set_xticks(x + width, ddf.index)
ax.legend(loc='upper left', ncols=3)
ax.set_ylim(0, 200000)


plt.show()
```
